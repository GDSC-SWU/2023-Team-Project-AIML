### 데이터를 Tensor화 하는 파일

```python
import pandas as pd
import torch
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('snunlp/KR-SBERT-V40K-klueNLI-augSTS')
```

```python
# data load
question = pd.read_excel('/content/questions.xlsx')
ques_db = pd.DataFrame(question)
ques_db.head()
```

```python
# 질문 행에 임베딩 수행
ques_db['embedding_vector'] = ques_db['질문 (Query)'].apply(model.encode)

# 파일 저장하기
ques_db.to_excel('question_producted_embedding.xlsx')
```

```python
# Tensor화 하기
data_to_tensor = torch.tensor(ques_db['embedding_vector'])

# tensor 파일만 저장하기 -> torch save 사용
torch.save(data_to_tensor, 'ques_tensor.pt')
```
