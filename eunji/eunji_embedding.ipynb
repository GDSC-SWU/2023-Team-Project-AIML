```python
import pandas as pd
from sentence_transformers import SentenceTransformer
import torch

#사전 학습 모델
model = SentenceTransformer('snunlp/KR-SBERT-V40K-klueNLI-augSTS')

df = pd.read_excel("Qusetion.xlsx")

#'질문 (Query)'열에 해당하는 부분에 대해서만 임베딩 진행하고 'embedding_vector' 열에 저장
df['embedding_vector'] = df['질문 (Query)'].apply(lambda x: model.encode(x))

#임베딩된 데이터를 엑셀 파일로 저장
df.to_excel("embedded_Questions.xlsx", index=False)

#임베딩 데이터를 torch.tensor()를 통해 텐서화한 후 .pt 파일로 저장
embedding_tensors = torch.tensor(df['embedding_vector'].tolist())
torch.save(embedding_tensors, 'embedding_data.pt')
```
